```python
#Importing Librarires
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Ignore Warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
#Reading the Csv file
df= pd.read_csv('CarPrice_Assignment.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#Copying the data
data= df.copy()
```

```python
#Displaying all the numeric values
numeric= ['int64','float64']
data1= data.select_dtypes(include=numeric)
data1
```

```python
data.info()
```

```python
#Displaying the percentage of Nan values
null_percentage= data.isna().sum().sum()*100/len(data)
null_percentage
```

```python
#Identify all the categorical values
#Drop all the categorical values
categoric=['object']
data2= data.select_dtypes(include=categoric)
data2.head(5)
```

```python
#Conversion of categorical values into numeric using Label encoder
#Data preprocessing

from sklearn import preprocessing
```

```python
#Applying LabelEncoder to data2

le= preprocessing.LabelEncoder()
data2= data2.apply(le.fit_transform)
data2.head()
```

```python
#Concat data2 with original data

data= data.drop(data2.columns, axis=1)
data= pd.concat([data, data2], axis=1)
data.head()
```

```python
# convert target variable income to categorical
data['price'] = data['price'].astype('int')
```

```python
# Importing train_test Split 

from sklearn.model_selection import train_test_split
```

```python
data.info()
```

```python
#Putting the feature variable 
X= data.drop('price',axis=1)
y= data['price']
```

```python
#Splitting the data into train and test 
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.30, random_state=99)
X_train.head()
```

```python
#Dropping all tye categorical to numerical columns
X_train= X_train.drop(['CarName','fueltype','doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber','fuelsystem'], axis=1)
X_train.head()
```

```python
X_train= X_train.drop(['aspiration', 'symboling','car_ID'], axis=1)
```

```python
#dropping all the not required features in X_test to match with model 

X_test= X_test.drop(['CarName','fueltype','doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber','fuelsystem','aspiration', 'symboling','car_ID'], axis=1)
```

```python
X_train.info()
```

```python
#Importing Decision tree classifier from sklearn library
from sklearn.tree import DecisionTreeClassifier

#fitting the decisison tree with default hyperparametres, apart from
# max depth is set to 5 so that we can plot and read the tree

dt_default= DecisionTreeClassifier(max_depth=5)
dt_default.fit(X_train, y_train)
```

```python
#lets check the evaluation matrix and accuracy

#Importing classification report and confusion matrix from sklearn metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

#Making Predictions
y_pred_default= dt_default.predict(X_test)

#printing classification report
print(classification_report(y_test, y_pred_default))
```

```python
#Printing Confusion matrix and accuracy
print (confusion_matrix(y_test, y_pred_default))
print (accuracy_score(y_test, y_pred_default))
```

```python
#Plotting thre Decision Tree
# Importing the required packages for visualization

from IPython.display import Image
from sklearn.externals.six import StringIO
from sklearn.tree import export_graphviz
import pydotplus, graphviz

#Putting features
features= list(data.columns[1:])
features
```

```python
#Plotting Tree with max_depth=3
dot_data= StringIO()
export_graphviz(dt_default, out_file= dot_data, feature_names= features, filled=True, rounded=True)
graph= pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

```python
#Tuning max_depth

# GridSearchCV to find optimal max_depth
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV


# specify number of folds for k-fold CV
n_folds = 2

# parameters to build the model on
parameters = {'max_depth': range(1, 40)}

# instantiate the model
dtree = DecisionTreeClassifier(criterion = "gini", 
                               random_state = 100)

# fit tree on training data
tree = GridSearchCV(dtree, parameters, 
                    cv=n_folds, 
                   scoring="accuracy")
tree.fit(X_train, y_train)
```

```python
# scores of GridSearch CV
scores = tree.cv_results_
pd.DataFrame(scores).head()
```

```python
# plotting accuracies with max_depth
plt.figure()
plt.plot(scores["param_max_depth"], 
         scores["mean_train_score"], 
         label="training accuracy")
plt.plot(scores["param_max_depth"], 
         scores["mean_test_score"], 
         label="test accuracy")
plt.xlabel("max_depth")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```
